```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Define the differential equation
def f(x, y):
    return x * np.exp(x) - y

# Exact solution function
def exact_solution(x):
    return x * np.exp(x) / 2

# Initial conditions
x0 = 1
y0 = np.exp(1)
x_end = 2
n_steps = 100
h = (x_end - x0) / n_steps

# Arrays to store the approximated solution and exact solution
x_values = np.linspace(x0, x_end, n_steps + 1)
y_approx = np.zeros(n_steps + 1)
y_exact = np.zeros(n_steps + 1)
errors = np.zeros(n_steps + 1)

# Initial values
y_approx[0] = y0
y_exact[0] = exact_solution(x0)

# Apply Euler's method
for i in range(n_steps):
    y_approx[i + 1] = y_approx[i] + h * f(x_values[i], y_approx[i])
    y_exact[i + 1] = exact_solution(x_values[i + 1])
    errors[i + 1] = abs(y_exact[i + 1] - y_approx[i + 1])

# Plot the exact solution and the approximated solution
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_exact, label='Exact Solution', color='blue')
plt.plot(x_values, y_approx, label='Euler Approximation', linestyle='--', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exact Solution vs Euler Approximation')
plt.legend()
plt.grid(True)
plt.show()

# Plot the errors
plt.figure(figsize=(10, 6))
plt.plot(x_values, errors, label='Error', color='purple')
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Error in Euler Approximation')
plt.legend()
plt.grid(True)
plt.show()
```
