```python
t = var('t')
x = sin(t)
y = cos(t)
plot([x, y], 0, 2*pi, color=['red', 'blue'])
```

```python
t = var('t')
x = vector([sin(t), cos(t)])
A = matrix([[0, 1], [-1, 0]])
x_der = x.diff(t)
A_x = A * x
sol = x_der == A_x
print(sol)
```

```python
x = var('x')
y = function('y')(x)
diff_eq = diff(y, x, 2) - 3*diff(y, x) + 2*y == x^2 + 3*x - 4
sol = desolve(diff_eq, y, ics=[0, 0, 0])
show(sol)

plot(sol, (x, -10, 5), ymin=-100, ymax=100).show()

y2 = sol.subs(x=2*pi).simplify()
y2p = diff(sol, x).subs(x=2*pi).simplify()
integ = integral(sol, x, 0, 1)
print(integ)
```

```python
var('x y')
f1 = x - y^2
f2 = x - y
J = jacobian([f1, f2], (x, y))
show(J)

eq = solve([f1 == 0, f2 == 0], x, y)
show(eq)

for p in eq:
    Jp = J.subs(p)
    show("Jacobian {}: {}").format(p, Jp)
    eigen = Jp.eigenvalues()
    show("Eigenvalues at {}: {}".format(p, eigenvalues))
```
